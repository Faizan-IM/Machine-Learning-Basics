# What is Machine Learning?

Machine Learning is the science (and art) of programming computers so they can *learn from data*.

Here is a slightly more general definition:
> [Machine Learning is the] field of study that gives computers the ability to learn
 without being explicitly programmed.
 —Arthur Samu, 1959

And a more engineering-oriented one:
>A computer program is said to learn from experience E with respect to some task T
 and some performance measure P, if its performance on T, as measured by P
 improves with experience 
.

# Why Use Machine Learning

Let's see a traditional approach of Machine Learning:



Consider we are writing a spam filter using traditional programming techniques:

1) First, we would consider what spam typically looks like. We might notice that some words and phrases (such as "4U," "credit card," "free," and "amazing") tend to come up a lot in the subject line. We will also notice a few other patterns in the sender's parts of the email.
2) We would write a detection algorithm for each of the patterns that we notice, and our program will flag the email as spam if these patterns were detected.
3) We would test the program and keep on repeating step 1 and 2 until it was good enough to launch.

Since the problem is difficult, our program will become a long list of complex rules-pretty hard to maintain. 

In contrast, a spam filter based on the **Machine Learning** technique automatically learns which words and phrases are good predictors of spam by detecting unusually frequent patterns of words in the spam, as shown below in the picture.


The program is much shorter, easier to maintain, and most likely more accurate.

Now let's think that spammers notice that all their emails containing "4U" are blocked so they start using "For U". The **Traditional Approach** will demand that we update our software with $for U$. This means we would keep writing new rules forever.

## Machine Learning and Humans

Machine Learning can help humans learn as shown below:



ML algorithms can be inspected to see what they have learned. For instance, once a spam filter has been trained on enough spam, it can easily be inspected to reveal a list of words and combination of words that it believes are the best predictors of spam. 

Sometimes they will reveal unsuspected co-relations or new trends, and thereby lead to a better understanding of the problem. Applying ML techniques to dig into large amounts of data can help discover patterns that were not imminently apparent. This is called ***data mining***.

### Summary

> Problems for which existing solutions require a lot of finetuning or long list of rules: one machine learning algorithm can often simplify the code and perform better than the traditional approaach.

> Complex problem for which using a traditional approach yeilds no good solution: the best machine learning techniques can perhaps find a solution.

> Fluctuating environments: a Machine Learning system can adapt to a new data.

> Getting insights about complex problems and large amount of data.
